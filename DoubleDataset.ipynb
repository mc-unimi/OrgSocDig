```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
tips1 = sns.load_dataset("tips")
tips1
```

```python
uomini = (tips1.sex == "Male")
donne = (tips1.sex == "Female")
```

```python
tips_uomo=tips1[uomini]
tips_uomo
```

```python
tips_donna=tips1[tips1.sex == "Female"]
tips_donna
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips_donna, color="red");
sns.scatterplot(x="total_bill", y="tip", data=tips_uomo, color="blue");
plt.show()
```

```python
type(sns.scatterplot(x="total_bill", y="tip", data=tips_donna, color="red"))
```

```python
type(sns.relplot(x="total_bill", y="tip", data=tips_donna, kind="scatter", color="red"))
```

```python
sns.relplot(x="total_bill", y="tip", data=tips_donna, kind="scatter", color="red");
sns.relplot(x="total_bill", y="tip", data=tips_uomo, kind="scatter", color="blue");
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips_donna, kind="line", color="red");
sns.relplot(x="total_bill", y="tip", data=tips_uomo, kind="line", color="blue");
plt.show()
```

```python
sns.lineplot(x="total_bill", y="tip", data=tips_donna, color="red");
sns.lineplot(x="total_bill", y="tip", data=tips_uomo, color="blue");
plt.show()
```
