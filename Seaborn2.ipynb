## BOXPLOT

```python
plt.figure(figsize=(10, 7))
colori={"yellow":"#e8d01a", "green":"#0a5c20"}
sns.boxplot(x="pickup_borough", y="distance", data=taxi, hue="color", palette=colori)
plt.show()
```

Usare i quartieri va bene perchÃ© abbiamo molti valori di distanza per quartiere.

```python
plt.figure(figsize=(10, 7))
colori={"yellow":"#e8d01a", "green":"#0a5c20"}
perc=taxi["tip"]/taxi["total"]
sns.boxplot(x="pickup_borough", y=perc*100, data=taxi, hue="color", palette=colori)
plt.show()
```

```python
plt.figure(figsize=(20, 7))
colori={"yellow":"#e8d01a", "green":"#0a5c20"}
giorno=taxi["pickup"].dt.day
sns.boxplot(x=giorno, y="total", data=taxi, hue="color", palette=colori)
plt.show()
```

## VIOLIN PLOT

```python
plt.figure(figsize=(10, 7))
colori={"yellow":"#e8d01a", "green":"#0a5c20"}
sns.violinplot(x="pickup_borough", y="distance", data=taxi, hue="color", palette=colori)
plt.show()
```

Se disegnamo un boxplot sovrapponendo uno stripplot vediamo per quali valori della y i punti sono piÃ¹ densi. La forma del violino rappresenta quella densitÃ 

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 7))
colori={"yellow":"#e8d01a", "green":"#0a5c20"}
sns.boxplot(x="pickup_borough", y="distance", data=taxi, hue="color", palette=colori)
sns.stripplot(x="pickup_borough", y="distance", data=taxi, hue="color", alpha=0.1, dodge=True, jitter=0.3)
plt.show()
```

```python
plt.figure(figsize=(20, 7))
colori={"yellow":"#e8d01a", "green":"#0a5c20"}
giorno=taxi["pickup"].dt.day
sns.violinplot(x=giorno, y="total", data=taxi, hue="color", palette=colori)
plt.show()
```

## HEATMAP

```python
giorno=taxi["pickup"].dt.day
taxi1=taxi.groupby(["pickup_borough",giorno])["total"].mean().reset_index()
taxi1
```

```python
taxi1=taxi1.fillna(0)
```

```python
taxi2=taxi1.pivot("pickup_borough", "pickup", "total")
taxi2
```

```python
taxi2=taxi2.fillna(0)
taxi2
```

```python
plt.figure(figsize=(20, 5))
ax=sns.heatmap(taxi2, cmap="viridis")
plt.yticks(rotation="horizontal")
plt.show()
```

## CLUSTERMAP

La rappresentazione ad albero gerarchico si chiama ***dendrogramma*** e per ogni livello associa gli elementi piÃ¹ simili tra loro

```python
g=sns.clustermap(taxi2, figsize=(10,6), cmap="icefire")
```

```python
giorno=taxi["pickup"].dt.day
taxi3=taxi.groupby(["pickup_borough",giorno])["distance"].mean().reset_index()
taxi3
```

```python
taxi3=taxi3.pivot("pickup_borough", "pickup", "distance").fillna(0)
taxi3
```

```python
g=sns.clustermap(taxi3, figsize=(10,6), cmap="icefire")
```

```python
giorno=taxi["pickup"].dt.day
taxi4=taxi.groupby(["pickup_zone",giorno])["tip"].mean().reset_index()
taxi4
```

```python
taxi4_1=taxi4.pivot("pickup_zone", "pickup", "tip").fillna(0)
taxi4_1
```

```python
g=sns.clustermap(taxi4_1, figsize=(20,60), cmap="icefire", dendrogram_ratio=(.2,.03))
```

```python
taxi4_2=taxi4.query('tip>4')
taxi4_2=taxi4_2.pivot("pickup_zone", "pickup", "tip").fillna(0)
taxi4_2
```

```python
g=sns.clustermap(taxi4_2, figsize=(20,30), cmap="icefire", dendrogram_ratio=(.2,.05))
```
