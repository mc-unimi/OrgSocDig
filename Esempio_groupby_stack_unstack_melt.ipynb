**Si legga il csv ds572_raccolta_differenziata_2010-16.csv dal titolo: Raccolta Differenziata (2010-2016), rilasciato dal Comune di Milano come Open Data (https://dati.comune.milano.it/dataset/ds572_raccolta-differenziata-2010-2016)**

```python
import numpy as np
import pandas as pd
```

```python
raccolta = pd.read_csv('ds572_raccolta_differenziata_2010-16.csv', sep =";")
raccolta
```

**1) Svolgere analisi preliminare del dataset (valori mancanti, coerenza dei tipi di dato, riniminare colonne, caratteristiche generali)**

```python
raccolta.isnull().any()
```

```python
raccolta['Rifiuti_differenziata'].isnull().sum()
```

```python
raccolta[raccolta['Rifiuti_differenziata'].isnull()]
```

Ho trovato che ci sono 5 valori nulli, nella colonna Rifiuti_differenziata, tutti relativi a Ingombranti a recupero. Potrei mettere 0 come valore, oppure ricavare un dataframe senza quelle righe. Scelgo questa ultima soluzione.

```python
raccolta=raccolta[~raccolta['Rifiuti_differenziata'].isnull()]
raccolta
```

```python
raccolta.dtypes
```

I tipi di dato sono coerenti

```python
raccolta.columns
```

```python
raccolta=raccolta.rename(columns={'anno_rilevamento_rifiuti': "anno", 
                        'Rifiuti_differenziata_tipologia': "tipo", 
                        'Rifiuti_differenziata': "quantita"})
raccolta
```

**AGGREGATE e TRANSFORM**

**Calcolare la somma per ogni tipologia di rifiuto su tutti gli anni della rilevazione e ordinare in modo decrescente**

```python
g=raccolta.groupby(['tipo'])[["quantita"]].aggregate('sum')
g.sort_values(['quantita'], ascending = False)
```

```python
raccolta.groupby(['anno'])[["quantita"]].sum()
```

```python
raccolta.groupby(['anno'])[['tipo','quantita']].min()
```

```python
raccolta.groupby(['anno'])[['tipo','quantita']].max()
```

```python
raccolta.groupby(['anno'])[['tipo','quantita']].aggregate(['min','max'])
```

```python
raccolta.groupby(['anno'])[['tipo','quantita']].agg(['min', 'max', 'mean'])
```

**Aggiungere una colonna percent con la percentuale del tipo di rifiuto rispetto il totale dei rifiuti del corrispondente anno** 

```python
pd.set_option('display.max_rows', 100)
raccolta6=raccolta
raccolta6['percent'] = 100 * raccolta['quantita'] / raccolta.groupby('anno')['quantita'].transform('sum')
raccolta6
```

```python
raccolta.groupby('tipo')['quantita'].transform('sum')
```

```python
raccolta.groupby('anno')['quantita'].transform('size')
```

```python
raccolta6.groupby(['anno'])[['percent','quantita']].agg(['min', 'max', 'mean'])
```

```python
raccolta6.groupby(['anno'])[['percent','quantita']].aggregate({'percent':['min', 'max'], 'quantita': 'mean'})
```

```python
raccolta6['num'] =raccolta.groupby('anno')['quantita'].transform('count')
```
