```python
import numpy as np
import pandas as pd
```

```python
emi = pd.read_csv("Emission.csv", sep=";")
emi
```

```python
pop = pd.read_csv("Population.csv", sep=";")
pop
```

**Prima proviamo con pd.concat()**

```python
pd.concat([emi, pop], join='outer')
```

```python
pd.concat([pop, emi], join='outer')
```

```python
pd.concat([emi, pop], join='inner')
```

**CHIARAMENTE IN QUESTO CASO LA CONCATENAZIONE È L'OPERAZIONE SBAGLIATA DA FARE**

**Vediamo la merge()**

```python
pd.merge(emi, pop, on='country', how='inner')
```

```python
pd.merge(emi, pop, on='country', how='outer')
```

```python
pd.merge(emi, pop, on='country', how='left')
```

```python
pd.merge(emi, pop, on='country', how='right')
```
