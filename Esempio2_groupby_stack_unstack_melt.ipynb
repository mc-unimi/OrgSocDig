```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
consumi = pd.read_csv('ds120_economia_spesa_media_mese_singole_voci_media_acquirenti_2007-2013.csv', 
                      sep =";", encoding='Latin-1')
consumi
```

**1) Svolgere analisi preliminare del dataset (valori mancanti, coerenza dei tipi di dato, caratteristiche generali)**

**ATTENZIONE: questo caso presenta diverse difficoltà nell'analisi preliminare. Qui non si hanno NaN ma dati con caratteri che indicano l'assenza di valori, quindi trovarli è più complicato. Oltre a questo abbiamo caratteri spuri che devono essere modificati per convertire i tipi di dati. Quindi, in questo caso sono riassunte quasi tutte le tipologie di azioni preliminari. In un testo d'esame non ci saranno tutte contemporaneamente, per questioni di tempo. Però occorre esercitarsi su tutti questi casi che sono frequenti in dataset reali.**

```python
consumi.isnull().any()
```

```python
consumi.dtypes
```

```python
consumi=consumi.rename(columns={"Spesa media mensile familiare (euro)": "Spesa_media", 
                        "Media calcolata rispetto agli effettivi acquirenti (euro)": "Spesa_media_eff", 
                        "Frequenza di acquisto": "Freq"})
consumi
```

**Le ultime tre colonna sono di tipo object invece che numerico perché contengono caratteri, virgola al posto di punto e %**

```python
consumi["Spesa_media"]=consumi["Spesa_media"].str.replace(",",".")
consumi["Spesa_media_eff"]=consumi["Spesa_media_eff"].str.replace(",",".")
consumi["Freq"]=consumi["Freq"].str.replace(",",".")
consumi["Freq"]=consumi["Freq"].str.replace("%","")
consumi
```

```python
consumi["Spesa_media"].astype(float)
```

**L'errore sta dicendo che ha trovato un valore nella colonna Spesa_media che ha - tra i caratteri e quindi non riesce a convertirlo. Cerchiamo quel valore**

**Cercare sulla doc di pandas la funzione contains**

```python
consumi["Spesa_media"].str.contains('-', regex=False).any()
```

```python
consumi[consumi["Spesa_media"].str.contains('-', regex=False)]
```

**L'abbiamo trovato, è la riga 599 che contiene dei - nei valori. La elminiamo. Il carattere ~ è il not logico**

```python
consumi=consumi[~consumi.Spesa_media.str.contains('-', regex=False)]
consumi
```

**Riproviamo con le conversioni di tipo di dati**

```python
consumi["Spesa_media"]=consumi["Spesa_media"].astype(float)
consumi["Spesa_media_eff"]=consumi["Spesa_media_eff"].astype(float)
consumi["Freq"]=consumi["Freq"].astype(float)
consumi.dtypes
```

```python
consumi.describe()
```

**Ok, abbiamo i dati con i tipi coerenti e le statistiche descrittive corrette**

## ANALISI

**Importante saper lavorare bene con group_by per calcolare risultati su gruppi di righe, funzioni aggregate/stack/unstack, pivot table per avere indici e data frame con visualizzazione estesa delle colonne (dataframe wide form), melt per creare dataframe di tipo chiave:valore (long form). Sono quasi tutte le stesse operazioni di trasformazioni che erano importanti nel corso di R**

**2a) Calcolare le medie delle variabili Spesa_media, Spesa_media_eff e Freq per ogni anno**

```python
consumi.groupby('Anno').mean().drop('Freq', axis=1)
```

**2b) Come prima ma per ogni Categoria**

```python
consumi.groupby('Categoria').mean().drop('Anno', axis=1)
```

**2c) Come prima ma per ogni Categoria divisa per anno**

```python
consumi.groupby(['Anno','Categoria']).mean()
```

**3a) Vogliamo la Spesa_media per Categoria come colonne (usare unstack)**

```python
consumi.groupby(['Anno','Categoria'])[['Spesa_media']].aggregate('mean').unstack()
```

**3b) Uguale a prima ma usando pivot table**

```python
consumi.pivot_table('Spesa_media', index='Anno', columns='Categoria')
```

**4) Trasformare la precedente eliminando l'indice su Anno e portarlo come colonna** 

```python
consumi2=consumi.pivot_table('Spesa_media', index='Anno', columns='Categoria').reset_index()
```

**guardo se ora Anno è effettivamente una colonna**

```python
list(consumi2.columns) 
```

```python
consumi2
```

**5) Trasformare il dataframe per avere le categorie come colonna chiave e i valori in una colonna valori associata**

```python
consumi_key=consumi2.melt(
               id_vars=["Anno"],  
               var_name="Categorie",             
               value_name="Valori")
consumi_key
```
