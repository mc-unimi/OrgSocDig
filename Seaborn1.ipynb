```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
taxi=sns.load_dataset("taxis")
taxi
```

```python
taxi.dtypes
```

```python
taxi['pickup']  = pd.to_datetime(taxi['pickup'])
taxi['dropoff']  = pd.to_datetime(taxi['dropoff'])
```

```python
gialli = (taxi.color == "yellow")
verdi = (taxi.color == "green")
```

```python
taxi_gialli=taxi[gialli]
taxi_gialli
```

```python
taxi_verdi=taxi[verdi]
taxi_verdi
```

## Differenza tra le funzioni scatterplot(), lineplot() e relplot()

```python
sns.scatterplot(x="total", y="tip", data=taxi_gialli, color="gold")
```

Per un elenco di colori predefiniti vedere qui: https://matplotlib.org/stable/gallery/color/named_colors.html

Guardando che tipo di oggetto produce la funzione scatterplot(), vediamo che è di tipo Subplot, semplificando, un grafico che viene composto aggiungendo le caratteristiche

```python
fig=sns.scatterplot(x="total", y="tip", data=taxi_gialli, color="yellow")
type(fig)
```

**Comporre il grafico per layer sovrapposti (simile a ggplot per R)**

Per la scelta dei singoli colori e delle palette di colori vedere il tutorial di Seaborn. 
<br> È possibile scegliere singoli colori dalla codifica esadecimale (hex) da pratici siti web fatti apposta, ad esempio cercando su Google "hex color picker" 

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x="total", y="tip", data=taxi_gialli, c=['#e8d01a'])
sns.scatterplot(x="total", y="tip", data=taxi_verdi, c=['#0a5c20'])
plt.show()
```

**Proviamo con la relplot()**

Con la relplot() si deve specificare che tipoo di grafico si vuole produrre (scatter o line), ma per il resto il risultato è analogo al precedente

```python
fig=sns.relplot(x="total", y="tip", data=taxi_gialli, color="gold", kind="scatter")
```

Vediamo di che tipo è l'oggetto prodotto: è di tipo FacetGrid, non Subplot come prima. Qui sta la principale differenza, in questo caso l'oggetto è già predisposto per una visualizzazzione a facet, che risulterà semplificata.

```python
type(fig)
```

Ci sono altre differenze.
Se cerco di specificare le dimensioni, come fatto prima con la scatterplot(), non va.

```python
plt.figure(figsize=(10, 7))
sns.relplot(x="total", y="tip", data=taxi_gialli, color="gold", kind="scatter")
plt.show()
```

Provando a comporre il grafico con livelli sovrapposti, il risultato è differente, perché questi sono oggetti di tipo FacetGrid, quindi già predefiniti per la visualizzazione a facet.

```python
sns.relplot(x="total", y="tip", data=taxi_gialli, color="gold", kind="scatter")
sns.relplot(x="total", y="tip", data=taxi_verdi, color="seagreen", kind="scatter")
```

**Torniamo a usare taxi, prima della divisione tra taxi gialli e taxi verdi**

Qui vogliamo che i colori assegnati da hue siano il giallo e il verde scelti prima con hex color picker (o colori specificati con nomi predefiniti). Definiamo l'associazione tra i VALORI "yellow" e "green" della colonna color del dataset taxi (ATTENZIONE, questi "yellow" e "green" sono valori del dataset, non colori) e ii colori che vogliamo usare noi. Poi si specifica la palette.

```python
plt.figure(figsize=(10, 7))
colori={"yellow":"#e8d01a", "green":"#0a5c20"}
sns.scatterplot(x="total", y="tip", data=taxi, hue="color", palette=colori)
plt.show()
```

```python
colori={"yellow":"#e8d01a", "green":"#0a5c20"}
sns.relplot(x="total", y="tip", data=taxi, hue="color", palette=colori, kind="scatter")
```

**Ora facciamo la visualizzazione per facets rispetto ai quartieri dove i passeggeri vengono caricati (colonna pickup_borough)**

Qui usiamo le funzioni scatterplot(). Bisogna specificare FacetGrid() e poi map()

```python
colori={"yellow":"#e8d01a", "green":"#0a5c20"}
g = sns.FacetGrid(taxi, col="pickup_borough", hue="color", palette=colori, col_wrap=2)
g.map(sns.scatterplot,"total","tip")
g.add_legend()
```

Qui usiamo la relplot(), che semplifica molto la definizione dei facets

```python
colori={"yellow":"#e8d01a", "green":"#0a5c20"}
sns.relplot(x="total", y="tip", data=taxi, 
            hue="color", palette=colori, kind="scatter",
           col="pickup_borough", col_wrap=2)
```

Usiamo due variabili per i facets: quartiere di carico dei passeggeri (pickup_borough) e quartiere di scarico dei passeggeri (dropoff_borough).
<br> Quanti facets verranno creati? 
<br>(numero di valori diversi di pickup_borough) X (numero di valori diversi di dropoff_borough)
<br> 4 X 4=16

```python
colori={"yellow":"#e8d01a", "green":"#0a5c20"}
sns.relplot(x="total", y="tip", data=taxi, 
            hue="color", palette=colori, kind="scatter",
           col="pickup_borough", row="dropoff_borough")
```

Questa sarebbe stata una scelta grafica poco buona, i facet sono poco chiari e molti non significativi. 

**Usando la lineplot() invece della scatterplot() la logica non cambia**

```python
sns.relplot(x="pickup", y="distance", data=taxi_verdi, kind="line")
```

```python
plt.subplots(figsize=(20, 7))
sns.lineplot(x="pickup", y="distance", data=taxi_verdi)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.subplots(figsize=(20, 7))
sns.lineplot(x="pickup", y="distance", data=taxi_gialli, color="gold")
sns.lineplot(x="pickup", y="distance", data=taxi_verdi, color="green")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.subplots(figsize=(20, 7))
sns.lineplot(x="pickup", y="distance", data=taxi, hue="color")
plt.xticks(rotation=90)
plt.show()
```

## Differenza tra le funzioni barplot(), stripplot() e catplot()

```python
sns.barplot(x="pickup_borough", y="tip", data=taxi_gialli, color="gold")
```

Guardando che tipo di oggetto produce la funzione barplot(), vediamo che è di tipo Subplot, semplificando, un grafico che viene composto aggiungendo le caratteristiche

```python
fig=sns.barplot(x="pickup_borough", y="tip", data=taxi_gialli, color="gold")
type(fig)
```

**Proviamo con la catplot()**

Con la catplot() si deve specificare che tipoo di grafico si vuole produrre (vedere il tutorial di Seaborn per i valori)

```python
fig=sns.catplot(x="pickup_borough", y="tip", data=taxi_gialli, color="gold", kind="bar")
```

```python
type(fig)
```

**Torniamo a usare taxi, prima della divisione tra taxi gialli e taxi verdi**

```python
plt.figure(figsize=(10, 7))
colori={"yellow":"#e8d01a", "green":"#0a5c20"}
sns.barplot(x="pickup_borough", y="tip", data=taxi, hue="color", palette=colori)
plt.show()
```

```python
colori={"yellow":"#e8d01a", "green":"#0a5c20"}
sns.catplot(x="pickup_borough", y="tip", data=taxi, hue="color", palette=colori, kind="bar")
```

**Ora facciamo la visualizzazione per facets rispetto ai quartieri dove i passeggeri vengono scaricati (colonna dropoff_borough)**

```python
colori={"yellow":"#e8d01a", "green":"#0a5c20"}
sns.catplot(x="pickup_borough", y="tip", data=taxi, 
            hue="color", palette=colori, kind="bar",
           col="dropoff_borough", col_wrap=2)
```

**Usando la stripplot() e la swarmplot()**

```python
plt.figure(figsize=(10, 7))
sns.stripplot(x="pickup_borough", y="tip", data=taxi_gialli, 
              palette=sns.color_palette("YlOrRd_r", 4))
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.swarmplot(x="pickup_borough", y="tip", data=taxi_verdi, 
              color="green")
plt.show()
```
