__Punto2: Estrarre i presidenti di altezza compresa tra 170 e 180 cm.__

```python
#versione senza operatore logico
presidenti170180 = data.loc[data['height(cm)'] >= 170]
presidenti170180 = presidenti170180.loc[presidenti170180['height(cm)'] <= 180]
presidenti170180
```

```python
#versione con operatore logico
data.loc[(data['height(cm)'] >= 170) & (data['height(cm)'] <= 180)]
```

__Punto3: Estrarre i presidenti che si chiamano George e sono più alti di 185 cm.__

```python
data.loc[(data['height(cm)'] >= 185) & (data['name'] == 'George')]
#perché?
```

```python
#capiamo meglio il funzionamento
#cosa restituisce la condizione logica?
data['height(cm)'] >= 185
```

```python
type(data['height(cm)'] >= 185)
```

La condizione logica restituisce una __serie__ di valori booleani

```python
#ora usiamo la funzione loc()
data.loc[data['height(cm)'] >= 185]
```

Quindi, la funzione __loc()__ applicata al __dataset data__ con la __serie di valori booleani come parametro__, produce il __subset di elementi di data corrispondenti a True__

```python
#ora proviamo la seconda condizione
data['name'] == 'George'
```

Nessun valore è True

```python
#proviamo con str.contains()
data['name'].str.contains('George')
```

Ora li abbiamo trovati. Quindi, l'operatore booleano == (così come gli altri <,>, <=,>=) non opera su stringhe. Servono funzioni specifiche per cercare i match su stringhe alfanumeriche (match, contains, etc.)

```python
#ora proviamo la combinazione delle due condizioni logiche
(data['height(cm)'] >= 185) & data['name'].str.match('George')
```

I valori booleani sono corretti

```python
data.loc[(data['height(cm)'] >= 185) & data['name'].str.match('George')]
```

__Punto 4: Estrarre i presidenti che si chiamano, come nome di battesimo, George, John, Arthur oppure James__

```python
#data
data.loc[data['name'].str.endswith('Arthur')]
```

```python
#con str.contains() o str.endswith() viene riconosciuto anche Arthur 
#(dettagli u espressioni regolari qui non li vediamo)
data.loc[data['name'].str.match('George') | 
         data['name'].str.match('John') | 
         data['name'].str.match('James') |
         data['name'].str.endswith('Arthur')]
```

```python
#senza Arthur come cognome
data.loc[data['name'].str.match('George') | 
         data['name'].str.match('John') | 
         data['name'].str.match('James') |
         data['name'].str.startswith('Arthur')]
```

```python
#senza Arthur come cognome
data.loc[data['name'].str.match('George') | 
         data['name'].str.match('John') | 
         data['name'].str.match('James') |
         data['name'].str.contains('Arthur')]
```
