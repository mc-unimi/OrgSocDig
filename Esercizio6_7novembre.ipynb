#### Lo scopo dell'analisi è quello di visualizzare la densità di popolazione per regione, prima nel dataset regioni poi nel dataset unito con ISTAT per vedere se il risultato è il medesimo. Poi si visualizzano le due regioni risultati più densamente popolate in un modo coerente e ordinato.

#### PUNTO PRIMO: DENSITA' ABITATIVA
* * *

+ **Visualizzare la densità di popolazione nel dataset regione**

```python
regioni = regioni.set_index('regione')
densita_pop = regioni['num_residenti']/regioni['superficie']
densita_pop
```

#### Le regioni con amggiore densità abitativa sono la Campania (427,97) e la Lombardia (419,39), che sono anche due delle regioni più popolose d'Italia.

+ **Unire i due dataset**

```python
regioni_pop=pd.merge(istat, regioni, left_on="Territorio", right_on="regione").drop('num_residenti', axis=1)
regioni_pop
```

+ **Utilizzare funzione melt per creare colonna Sesso e Popolazione**

```python
regioni_pop_tot=regioni_pop.melt(
               id_vars=["Territorio", "superficie","num_comuni", "num_provincie"],  
               var_name="Sesso",             
               value_name="Popolazione")
regioni_pop_tot
```

+ **Indicizzo secondo territorio**

```python
regioni_pop_tot1 = regioni_pop_tot.set_index(['Territorio'])
regioni_pop_tot1
```

+ **Selezionare solo la popolazione totale, non suddivisa tra maschi e femmine**

```python
regioni_pop_tot1 = regioni_pop_tot1.loc[regioni_pop_tot1['Sesso']=='Totale']
regioni_pop_tot1
```

+ **Visualizzare la densità abitativa**

```python
densita_pop2 = regioni_pop_tot1['Popolazione']/regioni_pop_tot1['superficie']
densita_pop2
```

#### Le regioni con la maggiore densità abitativa sono sempre la Campania (423,22) e la Lombardia (423,40), questa volta con valori differenti. Non essendo lo scarto troppo significativo, il motivo può essere un piccolo errore di conteggio (fermo restando che i dati ISTAT sono i più affidabili).

#### PUNTO SECONDO: VISUALIZZARE LE DUE REGIONI PIU' DENSAMENTE POPOLATE
* * *

+ **Selezionare le due regioni più densamente popolate**

```python
camp_lomb = regioni_pop_tot.loc[(regioni_pop_tot['Territorio']== 'Campania') | (regioni_pop_tot['Territorio']== 'Lombardia')]
camp_lomb
```

+ **Visuliazziamolo un po' meglio**

```python
camp_lomb_fin = camp_lomb.groupby(['Sesso', 'Territorio'])
```

```python
camp_lomb_fin.sum()
```

+ **In alternativa**

```python
camp_lomb1 = regioni_pop_tot.query('Territorio=="Campania" | Territorio=="Lombardia"')
camp_lomb1
```

```python
camp_lomb2 = camp_lomb1.groupby(['Territorio', 'Sesso'])
```

```python
camp_lomb2.sum()
```
